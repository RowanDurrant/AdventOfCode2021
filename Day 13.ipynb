```python
folds = [["x",655],["y",447],["x",327],["y",223],["x",163],["y",111],["x",81],["y",55],["x",40],["y",27],["y",13],["y",6]]
```

```python
foldLine = folds[0][1]
if folds[0][0] == "x":
    for j in range(len(df)):
        pos = df.loc[j]["x"]
        if pos > foldLine:
            df.at[j,"x"] = foldLine - (pos - foldLine)            
elif folds[0][0] == "y":
    for j in range(len(df)):
        pos = df.loc[j]["y"]
        if pos > foldLine:
            df.at[j,"y"] = foldLine - (pos - foldLine)
df2 = df.drop_duplicates()
len(df2)
```

```python
df = pd.read_csv("day12data.txt")
```

```python
for i in range(len(folds)):
    foldLine = folds[i][1]
    if folds[i][0] == "x":
        for j in range(len(df)):
            pos = df.loc[j]["x"]
            if pos > foldLine:
                df.at[j,"x"] = foldLine - (pos - foldLine)            
    elif folds[i][0] == "y":
        for j in range(len(df)):
            pos = df.loc[j]["y"]
            if pos > foldLine:
                df.at[j,"y"] = foldLine - (pos - foldLine)

df2 = df.drop_duplicates()
len(df2)
```

```python
plot = pd.DataFrame(columns = range(max(df2["x"])+1), index = range(max(df2["y"])+1))
plot[:][:] = "░"
```

```python
indices = list(df2.index.values)
for k in indices:
    X = df2.loc[k]["x"]
    Y = df2.loc[k]["y"]
    plot.at[Y,X] = "█"
```
